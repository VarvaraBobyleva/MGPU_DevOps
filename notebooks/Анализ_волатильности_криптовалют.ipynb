```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Блок 1: Загрузка данных

```python
df = pd.read_csv('/content/full_data__1__2021.csv')
df['timestamp'] = pd.to_datetime(df['timestamp'], unit='s')
df.set_index('timestamp', inplace=True)
df.dropna(axis=1, inplace=True)
```

```python
df
```

## Блок 2: Расчет волатильности

```python
# Простой расчет размаха цены внутри минуты
df['range'] = df['High'] - df['Low']

# Визуализация цены закрытия
df['Close'].plot(figsize=(12, 6), title='Цена Bitcoin (BTC) 2021')
plt.show()
```

## Блок 3: Вывод по волатильности

```python
print(f"Средняя минутная волатильность: {df['range'].mean():.2f} USD")
print(f"Максимальный скачок цены за минуту: {df['range'].max():.2f} USD")
```
